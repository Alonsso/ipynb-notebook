```
import matplotlib.pyplot as plt
import numpy as np
import scipy
import cv2
import pylab
from scipy import misc
from skimage.feature import local_binary_pattern
from skimage.feature import canny
from skimage import io
from skimage import color
```

```
img = cv2.imread('img1.png')

R = img[:, :, 0]
G = img[:, :, 1]
B = img[:, :, 2]
img_gray = R * 299. / 1000 + G * 587. / 1000 + B * 114. / 1000

img_gray = img_gray.astype('int32')
plt.imshow(img_gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```
plt.hist(img_gray.ravel(), bins='auto', color="grey")  # arguments are passed to np.histogram
plt.title("Histograma RGB ")
plt.show()
```

```
img_red = R.astype('int32')
plt.imshow(img_red, cmap = plt.get_cmap('Reds'))
plt.show()
```

```
plt.hist(img_red.ravel(), bins='auto', color="red")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Rojo")
plt.show()
```

```

img_green = G.astype('int32')
plt.imshow(img_green, cmap = plt.get_cmap('Greens'))
plt.show()
```

```

plt.hist(img_green.ravel(), bins='auto', color="green")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Verde")
plt.show()
```

```

img_blue = B.astype('int32')
plt.imshow(img_blue, cmap = plt.get_cmap('Blues'))
plt.show()
```

```

plt.hist(img_blue.ravel(), bins='auto', color="blue")  # arguments are passed to np.histogram
plt.title("Histograma del Componente Azul")
plt.show()
```

```

radius = 3
n_points = 8 *radius

METHOD = 'default'
lbp = local_binary_pattern(img_gray, n_points, radius, METHOD)
lbp = lbp.astype('int32')
plt.imshow(lbp,cmap = plt.get_cmap('gray'))
plt.show()
```

```
lbp = (lbp - lbp.min())/(lbp.max() - lbp.min()) 


plt.hist(lbp.ravel(), bins='auto', color="black")  # arguments are passed to np.histogram
plt.title("Histograma de textura")
plt.show()
```

```
dx = scipy.ndimage.sobel(img_gray, 0)  # derivada horizontal
dy = scipy.ndimage.sobel(img_gray, 1)  # derivada vertical
mag = np.hypot(dx, dy)  # magnitud
mag *= 255.0 / np.max(mag)  # normalizacin
plt.imshow(mag, cmap = plt.get_cmap('gray'))
plt.show()
```

```
plt.hist(mag.ravel(), bins='auto', color="black")  # arguments are passed to np.histogram
plt.title("Histograma de bordes")
plt.show()
```

```
(n, bins, patches) = plt.hist(mag.ravel(), bins=256, label='hst')
```
